```python
import random
import time

fail,big,small,count= 0,0,0,0

while True:
    count = count + 1
    
    a = random.randrange(1,7)
    b = random.randrange(1,7)
    c = random.randrange(1,7)

    if a+b+c == 3 or a+b+c==18:
        fail = fail + 1
    elif a+b+c<11 :
        small = small +1
    else:
        big = big + 1
    print('big확률:{0}, small확률:{1}'.format((big/count)*100,(small/count)*100))
    time.sleep(0.1)
```
